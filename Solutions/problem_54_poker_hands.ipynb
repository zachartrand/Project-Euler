## Problem 54
### Poker hands

<p>In the card game poker, a hand consists of five cards and are ranked, from lowest to highest, in the following way:</p>
<ul><li><b>High Card</b>: Highest value card.</li>
<li><b>One Pair</b>: Two cards of the same value.</li>
<li><b>Two Pairs</b>: Two different pairs.</li>
<li><b>Three of a Kind</b>: Three cards of the same value.</li>
<li><b>Straight</b>: All cards are consecutive values.</li>
<li><b>Flush</b>: All cards of the same suit.</li>
<li><b>Full House</b>: Three of a kind and a pair.</li>
<li><b>Four of a Kind</b>: Four cards of the same value.</li>
<li><b>Straight Flush</b>: All cards are consecutive values of same suit.</li>
<li><b>Royal Flush</b>: Ten, Jack, Queen, King, Ace, in same suit.</li>
</ul><p>The cards are valued in the order:<br />2, 3, 4, 5, 6, 7, 8, 9, 10, Jack, Queen, King, Ace.</p>
<p>If two players have the same ranked hands then the rank made up of the highest value wins; for example, a pair of eights beats a pair of fives (see example 1 below). But if two ranks tie, for example, both players have a pair of queens, then highest cards in each hand are compared (see example 4 below); if the highest cards tie then the next highest cards are compared, and so on.</p>

<p>The file, <a href="https://projecteuler.net/project/resources/p054_poker.txt">poker.txt</a>, contains one-thousand random hands dealt to two players. Each line of the file contains ten cards (separated by a single space): the first five are Player 1's cards and the last five are Player 2's cards. You can assume that all hands are valid (no invalid characters or repeated cards), each player's hand is in no specific order, and in each hand there is a clear winner.</p>
<p>How many hands does Player 1 win?</p>

### Solution

```python
RANKS = {f"{i}": i for i in range(2, 10)}
RANKS["T"] = 10
RANKS["J"] = 11
RANKS["Q"] = 12
RANKS["K"] = 13
RANKS["A"] = 14

RANK_SORT_KEY = lambda rank: RANKS[rank]


def get_suits(hand):
    suits = []
    for card in hand.split(" "):
        suits.append(card[-1])
    
    return suits


def get_ranks(hand):
    ranks = []
    for card in hand.split(" "):
        ranks.append(card[0])
    ranks.sort(key=RANK_SORT_KEY)
    
    return ranks


def get_count(values):
    counts = {}
    for value in values:
        counts.setdefault(value, 0)
        counts[value] += 1
    
    return counts


def is_straight(ranks):
    if ranks == ["2", "3", "4", "5", "A"]:
        return True
    
    for i, rank in enumerate(ranks):
        if i > 0 and RANKS[rank] != RANKS[ranks[i-1]] + 1:
            return False

    return True


def is_flush(suits):
    suits_set = set(suits)
    if len(suits_set) == 1:
        return True
    
    return False


def is_royal_flush(suits, ranks):
    if is_flush(suits) and is_straight(ranks):
        if ranks == ["T", "J", "Q", "K", "A"]:
            return True
    
    return False


def is_straight_flush(suits, ranks):
    if (
            is_flush(suits)
            and is_straight(ranks)
            and not is_royal_flush(suits, ranks)):
        return True
    
    return False


def is_full_house(rank_count):
    counts = [count for count in rank_count.values()]
    if len(counts) == 2:
        counts.sort()
        if counts == [2, 3]:
            return True
    
    return False


def is_four_of_a_kind(rank_count):
    if 4 in rank_count.values():
        return True
    
    return False


def is_three_of_a_kind(rank_count):
    if 3 in rank_count.values() and len(rank_count.keys()) == 3:
        return True
    
    return False


def is_two_pair(rank_count):
    counts = [count for count in rank_count.values()]
    if len(counts) == 3:
        counts.sort()
        if counts == [1, 2, 2]:
            return True
    
    return False


def is_pair(rank_count):
    counts = [count for count in rank_count.values()]
    if len(counts) == 4:
        return True
    
    return False


def is_high_card(rank_count):
    counts = [count for count in rank_count.values()]
    if len(counts) == 5:
        return True
    
    return False


def score_rank(rank):
    value = RANKS[rank]
    if value < 10:
        return f"0{value}"
    else:
        return str(value)


def score_hand(hand):
    ranks = get_ranks(hand)
    suits = get_suits(hand)
    rank_count = get_count(ranks)
    if is_pair(rank_count):
        hand_value = ["1"]
        for rank, count in rank_count.items():
            if count == 2:
                hand_value.append(score_rank(rank))
                ranks.remove(rank)
                ranks.remove(rank)
        for rank in reversed(ranks):
            hand_value.append(score_rank(rank))
        hand_value.append("00")
        
    elif is_two_pair(rank_count):
        hand_value = ["2"]
        pair_ranks = []
        for rank, count in rank_count.items():
            if count == 2:
                pair_ranks.append(rank)
        pair_ranks.sort(key=RANK_SORT_KEY, reverse=True)
        for rank in pair_ranks:
            hand_value.append(score_rank(rank))
            ranks.remove(rank)
            ranks.remove(rank)
        for rank in reversed(ranks):
            hand_value.append(score_rank(rank))
        hand_value.append("0000")
    
    elif is_three_of_a_kind(rank_count):
        hand_value = ["3"]
        for rank, count in rank_count.items():
            if count == 3:
                hand_value.append(score_rank(rank))
                for _ in range(3):
                    ranks.remove(rank)
        for rank in reversed(ranks):
            hand_value.append(score_rank(rank))
        hand_value.append("0000")
    
    elif is_royal_flush(suits, ranks):
        return 90000000000
    
    elif is_straight_flush(suits, ranks):
        hand_value = ["8"]
        hand_value.append(score_rank(ranks.pop()))
        hand_value.append("00000000")
    
    elif is_straight(ranks):
        hand_value = ["4"]
        if ranks == ["2", "3", "4", "5", "A"]:
            hand_value.append("0500000000")
        else:
            hand_value.append(score_rank(ranks.pop()))
            hand_value.append("00000000")
                
    elif is_flush(suits):
        hand_value = ["5"]
        for rank in reversed(ranks):
            hand_value.append(score_rank(rank))
        
    elif is_full_house(rank_count):
        hand_value = ["6"]
        for rank, count in rank_count.items():
            if count == 3:
                high_rank = rank
            else:
                low_rank = rank
        hand_value.append(score_rank(high_rank))
        hand_value.append(score_rank(low_rank))
        hand_value.append("000000")
        
    elif is_four_of_a_kind(rank_count):
        hand_value = ["7"]
        for rank, count in rank_count.items():
            if count == 4:
                high_rank = rank
            else:
                low_rank = rank
        hand_value.append(score_rank(high_rank))
        hand_value.append(score_rank(low_rank))
        hand_value.append("000000")
    
    elif is_high_card(rank_count):
        hand_value = ["0"]
        for rank in reversed(ranks):
            hand_value.append(score_rank(rank))
        
    score = "".join(hand_value)
    
    return int(score)
```

```python
POKER_FILE = "poker.txt"
with open(POKER_FILE) as file:
    poker_hands = file.read()

poker_hands = poker_hands.split("\n")
hand_1_wins = 0
for i, row in enumerate(poker_hands):
    hand_1 = row[:14]
    hand_2 = row[15:]
    hand_1_score = score_hand(hand_1)
    hand_2_score = score_hand(hand_2)
    
    if hand_1_score > hand_2_score:
        hand_1_wins += 1
```
